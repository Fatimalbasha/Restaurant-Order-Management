```python
# Function to print a welcome message in the beginning of the code
def welcome_message():
    print("Welcome to Ratatouille Restaurant!")
    print("You can order food, see the total, and get a receipt.")
    print("Choose items from each section. Type 'Next' to move to the next section or 'Done' to finish.")



# The menu of the restaurant including the sections and the items
menus = {
            "Appetizers": {
                             "Caesar Salad": 17,
                             "Fattoush Salad": 12,
                             "Hummus": 10,
                             "Grape Leaves": 15,
                             "Falafel": 11
                          },

            "Maindish": {
                             "Pasta": 20,
                             "Beef Burger": 16,
                             "Chicken Burger": 14,
                             "Dynamite Chicken": 25,
                             "Chicken Strips": 17,
                             "Pizza": 22
                        },


            "Side dishes": {
                             "Ice Cream": 4,
                             "Chocolate Cookies": 7,
                             "Potato Wedges": 8,
                             "Onion Rings": 8,
                             "Mozzarella Sticks": 12
                           },

            "Drinks": {
                             "Apple Juice": 6,
                             "Orange Juice": 6,
                             "Lemon Juice": 6,
                             "Coca Cola": 8,
                             "Mirinda": 8,
                             "Ice Tea": 9,
                             "Water": 2
                        }
         }



# Function to print the menu
def show_menu():
    for section, items in menus.items():
        print(f"\n {section}:")
        for item, price in items.items():
            print(f"  {item}: {str(price)} SAR")


# Function to take the order from the user and the quantity
def take_order(menus):
    order = {}
    normalized_menu = {}

    for section, items in menus.items():
        for item in items:
            normalized_menu[item.lower()] = item

    for section, items in menus.items():
        print(f"\n {section}:")

        while True:
            choice = input("What would you like to order from this section? (Type 'Next' to skip or 'Done' to finish): ").lower()

            if choice.lower() == 'next':
                break
            elif choice.lower() == 'done':
                return order
            elif choice in normalized_menu and normalized_menu[choice] in items:
                original_choice = normalized_menu[choice]
                quantity = int(input(f"How many {original_choice} would you like? "))
                order[original_choice] = quantity
            else:
                print("Invalid choice please try again.")
    return order


# Function to calculate the total bill
def calculate_total(order, menus):
    total = 0

    for item, quantity in order.items():
        for section in menus:
            if item in menus[section]:
                value = menus[section][item] * quantity
                total = total + value

    return total


# Function to generate a simple receipt
def generate_receipt(order, menus, total):
    print("\n\n\nReceipt: ")
    for item, quantity in order.items():
      for section in menus:
            if item in menus[section]:
                price = menus[section][item]
                print(f"{item}: {quantity} x {price} SAR = {price * quantity} SAR")
    print(f"Total: {total} SAR")


# This function is for user feedback and rating
def get_feedback():
    print("\nWe hope you enjoyed your meal! We'd love to hear your feedback.")

    experience_feedback = input("Please share your experience with us: ")

    service_rating = int(input("On a scale of 1 to 10, how would you rate our service? "))

    if service_rating < 5:
        print("We will work on improving our service.")
    else:
        print("We are glad to have met your expectations :) ")


# Main program
welcome_message()
show_menu()

order = take_order(menus)
total = calculate_total(order, menus)
generate_receipt(order, menus, total)

print(f"\n\n\nTotal Bill: \nYour total is: {str(total)} SAR \nThank you for visiting ^_^ ")
get_feedback()

```
